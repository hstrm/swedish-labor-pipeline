```python
import requests
import pandas as pd
import duckdb
import sys
```

```python
url = "https://api.scb.se/OV0104/v1/doris/sv/ssd/START/BE/BE0101/BE0101A/BefolkningNy"
response = requests.get(url)
metadata = response.json()

for variable in metadata["variables"]:
    print(f"\n{variable['code']} — {variable['text']}")
    for kod, text in zip(variable["values"], variable["valueTexts"]):
        print(f"  {kod}  {text}")
```
